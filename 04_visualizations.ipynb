```python
# -----------------------------------------
# Healthcare Readmission Analytics Project
# Visualizations & Business Insights
# -----------------------------------------

from pyspark.sql.functions import *

# Gold table
gold_table = "workspace.gold.healthcare_kpis_gold"

# Silver table
silver_table = "workspace.silver.diabetic_data_silver"
```

```python
# Read tables

df_gold = spark.table(gold_table)

df_silver = spark.table(silver_table)
```

Readmission Rate by Age

```python
# Readmission rate by age

readmission_age_df = (
    df_silver
    .groupBy("age")
    .agg(
        round(
            avg("readmitted_flag") * 100,
            2
        ).alias("readmission_rate")
    )
    .orderBy("age")
)

display(readmission_age_df)
```

Patients by Gender

```python
# Patients by gender

gender_vis_df = (
    df_silver
    .groupBy("gender")
    .count()
)

display(gender_vis_df)
```

Medication Usage Distribution

```python
# Medication level distribution

medication_vis_df = (
    df_silver
    .groupBy("medication_level")
    .count()
)

display(medication_vis_df)
```

Average Hospital Stay by Category

```python
# Average stay by hospital category

stay_vis_df = (
    df_silver
    .groupBy("hospital_stay_category")
    .agg(
        round(
            avg("days_in_hospital"),
            2
        ).alias("avg_stay")
    )
)

display(stay_vis_df)
```
